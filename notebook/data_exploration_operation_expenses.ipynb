```python
data.columns = ['operational_hours_costs','1995','1996','1997','1998','1999','2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020']
```

```python
def fleet_operation_expenses(data):
    fleet_costs_df = data.drop(columns = ['operational_hours_costs'],axis=1)
    fleet_costs_df =  fleet_costs_df.iloc[-1]
    fleet_costs_df = fleet_costs_df.astype(int)
    fleet_costs_df=fleet_costs_df.T
    sns.set()
    plt.tight_layout()
    plt.title(" Aircrafts in Fleet - 1995 to 2020")
    plt.ylabel("Dollars spent in 10 million")
    plt.xlabel("years in operation")
    ax = fleet_costs_df.plot(figsize=(10, 8))    
```

```python
def aircraft_in_fleet(data):
    fleet_costs_df = data.drop(columns = ['operational_hours_costs'],axis=1)
    fleet_costs_df = fleet_costs_df.iloc[0]
    fleet_costs_df = fleet_costs_df.astype(int)
    fleet_costs_df = fleet_costs_df.T
    sns.set()
    plt.tight_layout()
    plt.title("Overall fleet operation cost for American airlines - 1995 to 2020")
    plt.ylabel("Number of flights")
    plt.xlabel("years in operation")
    ax = fleet_costs_df.plot(figsize=(10, 8))  
```

```python
def seats_per_departure(data):
    fleet_costs_df = data.drop(columns = ['operational_hours_costs'],axis=1)
    fleet_costs_df = fleet_costs_df.iloc[1]
    fleet_costs_df = fleet_costs_df.astype(int)
    fleet_costs_df = fleet_costs_df.T
    sns.set()
    plt.tight_layout()
    plt.title("Total Seats per departure - 1995 to 2020")
    plt.ylabel("Number of seats")
    plt.xlabel("years in operation")
    ax = fleet_costs_df.plot(figsize=(10, 8))      
    
```

```python
def sum_of_departures(data):
    fleet_costs_df = data.drop(columns = ['operational_hours_costs'],axis=1)
    fleet_costs_df = fleet_costs_df.iloc[6]
    fleet_costs_df = fleet_costs_df.astype(int)
    fleet_costs_df = fleet_costs_df.T
    sns.set()
    plt.tight_layout()
    plt.title("Overall departures per year - 1995 to 2020")
    plt.ylabel("Number of departures")
    plt.xlabel("years in operation")
    ax = fleet_costs_df.plot(figsize=(10, 8))  
```

```python
def sum_of_rpm(data):
    fleet_costs_df = data.drop(columns = ['operational_hours_costs'],axis=1)
    fleet_costs_df = fleet_costs_df.iloc[7]
    fleet_costs_df = fleet_costs_df.astype(int)
    fleet_costs_df = fleet_costs_df.T
    sns.set()
    plt.tight_layout()
    plt.title(" Revenue passenger Mile- 1995 to 2020")
    plt.ylabel("Number of Miles")
    plt.xlabel("years in operation")
    ax = fleet_costs_df.plot(figsize=(10, 8)) 
```

```python
def sum_of_asm(data):
    fleet_costs_df = data.drop(columns = ['operational_hours_costs'],axis=1)
    fleet_costs_df = fleet_costs_df.iloc[8]
    fleet_costs_df = fleet_costs_df.astype(int)
    fleet_costs_df = fleet_costs_df.T
    sns.set()
    plt.tight_layout()
    plt.title(" Available seat Miles - 1995 to 2020")
    plt.ylabel("Number of Miles")
    plt.xlabel("years in operation")
    ax = fleet_costs_df.plot(figsize=(10, 8))    
```
